### Question 3

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import math
import random
```

#### part a)

```python
X = np.linspace(0, 1, 100)
X = X.reshape(-1,1)
Y = np.sqrt(X**3) # my f(x) function

# making the tree
MSE = DecisionTreeRegressor(criterion = 'squared_error', max_leaf_nodes=6) 
MAE = DecisionTreeRegressor(criterion = 'absolute_error', max_leaf_nodes=6)

# fitting the tree to variables
MSE.fit(X, Y)
MAE.fit(X, Y)

# predicting labels
MSEpredict = MSE.predict(X)
MAEpredict = MAE.predict(X)

#plotting tree
plt.figure()
plt.scatter(X, Y, s=10, label = 'f(X)')

plt.plot(X, MSEpredict, color='red', label = 'MSE Predictor')
plt.plot(X, MAEpredict, color='green', label = 'MAE Predictor')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Decision Tree Regression with MSE and MAE')
plt.legend()
plt.show()
```

### part b)

In my decision tree regression, you can that the accuracy of the predictors get better with higher values of x. At low values of x, the MAE predictor does better than MSE, as the MSE is more sensitive to outliers due to it squaring the values. From x = 0 to around 0.7, the MSE predictor is slightly higher and more off than the MAE predictor due to this sensitivity from squaring. After 0.7, you can see both starting to adjust to the pattern of the data. Both regressors also slightly underfit the data too, regardless of using MAE or MSE. 

## Question 4

```python
import numpy as np
import math
from math import pi
import random
import matplotlib.pyplot as plt

# generate data
x = np.linspace(0,1,100)
y1 = np.sqrt(x**3) + np.random.normal(scale = 0.05, size=100) # my f(x) and added noise at a scale of 0.05

# This function interates over each term of the polynomial and accumulating the results in y
def eval_poly(c,x):
    y = np.zeros_like(x)
    for coeff in c:
        y = y * x + coeff
    return y

# Using polyfit for my overfit and underfit. Specified the degree of the polynomial to be 10 for the overfitting, and 0.5 for the underfitting. 
fit = np.polyfit(x,y1,10)
fit2 = np.polyfit(x,y1,0.5)
overfit = eval_poly(fit,x)
underfit = eval_poly(fit2,x)

# Plottin original data
plt.plot(x,y1, 'o', label='Data')

# Overfitting example and plot
plt.plot(x,overfit, '-', label='Overfitting')

# Underfitting example and plot
plt.plot(x,underfit,'-', label='Underfitting')

#Combine plots and add features
plt.title('Overfitting and Underfitting on Polynomial Function')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# Checking Error MSE
# I did the mean of the squared differences between y1 and the predicted value (which was overfit or underfit line).

#Overfitting MSE 
mseoverfit = np.mean((y1-overfit)**2)
print("MSE Overfit = ",mseoverfit)

#Underfitting MSE
mseunderfit = np.mean((y1-underfit)**2)
print("MSE Underfit = ",mseunderfit)
```

From my results it is clear that the overfitting is more accurate, as there is a lower mean squared error than the underfitting. However, it is interesting to note that my overfitted line is moving with the noise in the data, and not actually matching to the square root function I am actually trying to plot. This shows some degree of "better accuracy", but since the overfit graph is just tracking the noise, it is not actually accurate and has high variance.
